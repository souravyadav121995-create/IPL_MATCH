```python
# IPL DATA ANALYSIS

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import streamlit as st
```

```python

scorecard = pd.read_csv("matches.csv")
```

```python
scorecard.head()

st.title("IPL Data Analysis")
st.dataframe(scorecard.head())
plt.show()
```

```python
scorecard['winner'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='winner', data=scorecard, order=scorecard['winner'].value_counts().index)
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Number of Matches Won')
plt.ylabel('Teams')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=scorecard['winner'].value_counts().index, y=scorecard['winner'].value_counts().values, palette='viridis')
plt.title('Number of Matches Won by Each Team in IPL')
plt.xlabel('Teams')
plt.ylabel('Number of Matches Won')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='season', palette='magma')
plt.title('Number of Matches per Season in IPL')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='city', palette='coolwarm', order=scorecard['city'].value_counts().index)
plt.title('Number of Matches Played in Each City')
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='toss_winner', palette='Set2', order=scorecard['toss_winner'].value_counts().index)
plt.title('Number of Tosses Won by Each Team')
plt.xlabel('Teams')
plt.ylabel('Number of Tosses Won')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='result', palette='pastel')
plt.title('Match Results Distribution in IPL')
plt.xlabel('Result Type')
plt.ylabel('Number of Matches')
plt.show()
```

```python

plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='venue', palette='cubehelix', order=scorecard['venue'].value_counts().index)
plt.title('Number of Matches Played at Each Venue')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()

```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='umpire1', palette='Set3', order=scorecard['umpire1'].value_counts().index)
plt.title('Number of Matches Officiated by Each Umpire')
plt.xlabel('Umpire 1')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='umpire2', palette='Set3', order=scorecard['umpire2'].value_counts().index)
plt.title('Number of Matches Officiated by Each Umpire')
plt.xlabel('Umpire 2')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()

```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='player_of_match', palette='Spectral', order=scorecard['player_of_match'].value_counts().index[:20])
plt.title('Top 20 Players with Most Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.xticks(rotation=90)
plt.show()


```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='toss_decision', palette='Accent')
plt.title('Toss Decisions in IPL Matches')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Matches')
plt.show()
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='dl_applied', palette='Dark2')
plt.title('Number of Matches where D/L Method was Applied')

```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='season', hue='toss_decision', palette='Set1')
plt.title('Toss Decisions per Season in IPL Matches')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='season', hue='result', palette='Set2')
plt.title('Match Results per Season in IPL')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.show()

```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='winner', hue='toss_winner', palette='Set3', order=scorecard['winner'].value_counts().index)
plt.title('Matches Won by Teams vs Toss Winners')
plt.xlabel('Teams')
plt.ylabel('Number of Matches')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='venue', hue='winner', palette='Set1', order=scorecard['venue'].value_counts().index)
plt.title('Matches Won by Teams at Different Venues')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=scorecard, x='dl_applied', palette='Dark2')
plt.title('Number of Matches where D/L Method was Applied')
plt.xlabel('D/L Applied')
plt.ylabel('Number of Matches')
plt.show()
```
